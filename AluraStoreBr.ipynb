### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# prompt: Análise do faturamento

import pandas as pd
total_faturamento_loja1 = loja['Preço'].sum()
print(f'Total de faturamento da Loja 1: R$ {total_faturamento_loja1:.2f}')

total_faturamento_loja2 = loja2['Preço'].sum()
print(f'Total de faturamento da Loja 2: R$ {total_faturamento_loja2:.2f}')

total_faturamento_loja3 = loja3['Preço'].sum()
print(f'Total de faturamento da Loja 3: R$ {total_faturamento_loja3:.2f}')

total_faturamento_loja4 = loja4['Preço'].sum()
print(f'Total de faturamento da Loja 4: R$ {total_faturamento_loja4:.2f}')

# Concatenar todos os dataframes em um só para análise consolidada
df_todas_lojas = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

total_faturamento_geral = df_todas_lojas['Preço'].sum()
print(f'\nTotal de faturamento geral de todas as lojas: R$ {total_faturamento_geral:.2f}')
```

# 2. Vendas por Categoria


```python
# prompt: Vendas por Categoria

vendas_por_categoria = df_todas_lojas.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)
print("\nVendas por Categoria:")
vendas_por_categoria
```

# 3. Média de Avaliação das Lojas

```python
# prompt: Média de Avaliação das Lojas

# Calcula a média das avaliações para cada loja
media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print(f'\nMédia de avaliação da Loja 1: {media_avaliacao_loja1:.2f}')
print(f'Média de avaliação da Loja 2: {media_avaliacao_loja2:.2f}')
print(f'Média de avaliação da Loja 3: {media_avaliacao_loja3:.2f}')
print(f'Média de avaliação da Loja 4: {media_avaliacao_loja4:.2f}')

# Calcula a média geral de avaliação de todas as lojas
media_avaliacao_geral = df_todas_lojas['Avaliação da compra'].mean()
print(f'\nMédia geral de avaliação de todas as lojas: {media_avaliacao_geral:.2f}')
```

# 4. Produtos Mais e Menos Vendidos

```python
# prompt: Produtos Mais e Menos Vendidos

# Conta a quantidade de cada produto vendido
vendas_por_produto = df_todas_lojas['Produto'].value_counts()

# Produtos mais vendidos (Top 10)
produtos_mais_vendidos = vendas_por_produto.head(10)
print("\nProdutos Mais Vendidos:")
print(produtos_mais_vendidos)

# Produtos menos vendidos (Bottom 10)
produtos_menos_vendidos = vendas_por_produto.tail(10)
print("\nProdutos Menos Vendidos:")
produtos_menos_vendidos
```

# 5. Frete Médio por Loja


```python
# prompt: Frete Médio por Loja

import pandas as pd
# Adiciona uma coluna para identificar a loja em cada dataframe antes de concatenar
loja['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

# Concatenar todos os dataframes em um só para análise consolidada
df_todas_lojas = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

# Calcula o frete médio por loja
frete_medio_por_loja = df_todas_lojas.groupby('Loja')['Frete'].mean()

print("\nFrete Médio por Loja:")
frete_medio_por_loja
```
